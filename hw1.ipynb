```python
# read in the iris data
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()

# create X (features) and y (response)
X = iris['data']
y = iris['target']
```

```python
data=pd.DataFrame(X,columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data['Species']= y
data.head()
```

**Evaluation procedure - Train and test on the entire dataset**

```python
# Use Logistic regression
import sklearn.linear_model as lm
regr = lm.LinearRegression()
X = data[["Petal length","Petal Width","Sepal Length","Sepal Width"]].values
y = data[['Species']].values
regr.fit(X,y)

data['LogRegPredict'] = regr.predict(X)
data

#prediction = pd.DataFrame({'Actual':[y], 'Predict':[y_pred]})
#prediction.head()
#prediction = pd.DataFrame([y,y_pred],columns=["Actual","Predict"])
#prediction.head()
```

```python
#Making truth table
guessed = data['LogRegPredict'].round() == data['Species']
actual = data['Species'] == data['Species']

cmLogReg = pd.crosstab(guessed, actual, rownames = ['guessed'], colnames = ['actual'])
cmLogReg
```

```python
# Compute accuracy for evaluating the model
accuracy = cmLogReg[1][1]/(cmLogReg[1][0]+cmLogReg[1][1])

print('Accuracy: ', accuracy)
```

#### Part II (Classification)

```python
# Use KNN 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X, y.ravel())

data['KNNPredict'] = knn.predict(X)
data
```

```python
#Truth table
guessed = data['KNNPredict'] == data['Species']
actual = data['Species'] == data['Species']

cmKNN = pd.crosstab(guessed, actual, rownames = ['guessed'], colnames = ['actual'])
cmKNN
```

```python
# Compute accuracy for evaluating the model
accuracy = cmKNN[1][1]/(cmKNN[1][0]+cmKNN[1][1])

print('Accuracy: ', accuracy)
```

#### Part III (Comparing results)
