```python
# load the data into panda dataframe
# load the data into pandas dataframe
# make sure the data is in json format
# TODO
import pandas as pd
df = pd.read_json('reviews_Musical_Instruments_5.json', lines = True)
df
```

```python
# Convert all the review text into a long string and print its length
# TODO
rtext = ""
for text in df.reviewText:
    rtext = rtext + " " + text
print(len(rtext))
```

```python
# Load the tokenizer
# tokenize into sentences
# TODO
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')

sent_tolkens = sent_tokenize(rtext)
sent_tolkens
```

```python
# clean and split sentences into words
# write a function to do this
# TODO

#import re library
import re
#import stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

#removes special characters and makes all strings lowercase
def removeSpecialCharacters(string):
    strip_special_chars = re.compile("[^A-Za-z]+")
    string = re.sub(strip_special_chars, " ", string)
    return string.strip().split()
#removes stop words 
def removeStopWords(arr):
    stopwordless = []
    for w in arr:
        if w.lower() not in stop_words:
            stopwordless.append(w)
    return stopwordless
```

```python
# might want to compute the word and their frequency and store somewhere
# you will need it later in the code
# TODO

#remove special characters and remove stop words
cleanedwords = removeStopWords(removeSpecialCharacters(rtext))

#import counter library
from collections import Counter

#counts the frequency of each word and prints them in descending order
counts = Counter(cleanedwords)
print(counts)
```

```python
# clean each raw sentences and build the list of sentences
# ideally you are calling the funtion on each sentence you tokenized earlier
# TODO

#create an array of sentence arrays that contain words that are cleaned
cleanedsentences = []
for x in sent_tolkens:
    cleanedsentences.append(removeStopWords(removeSpecialCharacters(x)))
cleanedsentences
```

```python
# check a sentence by printing it
# TODO

#joins together array of words into a single string sentence
(" ").join(cleanedsentences[3])
```

```python
# Build your word2vec model
# TODO
model = Word2Vec(cleanedsentences, min_count=1)
```

```python
# print top 10 words based on term frequency (make sure they are not stop-words)
# TODO

mostcommon = []
# use counts library to print out top 10 most common words and adds the words to a list
print 'Most common:'
for word, count in counts.most_common(10):
    print '%s: %7d' % (word, count)
    mostcommon.append(word)
```

```python
# print most similar top 5 words for each of the above words
# TODO
for x in mostcommon:
    print(model.most_similar(positive=[model[x]], topn = 5))
```
